```python
b = np.array([2])
s = np.array([1])
p = np.array([])
n = 50
for i in range(n):
    p = np.append(p, b[i] * s[i])
    b = np.append(b, 2 * b[i])
    s = np.append(s, np.sqrt((1.0 - np.sqrt(1 - s[i]**2)) / 2.0))
plt.title("Valori di p")
plt.plot(range(n), p, "r")
plt.hlines(math.pi, 0, n, colors="b", linestyles="--")
plt.show()

e = lambda p: np.abs(p - math.pi) / np.abs(math.pi)

plt.title("Errore relativo")
plt.plot(range(n), e(p), "r")
plt.show()

# l'algoritmo non Ã¨ numericamente stabile in quanto s -> 0 portando quindi 1.0 - sqrt(1 - s^2) = 0 per via dello spacing
# e di conseguenza tutti i futuri valori di s e quindi p.
# per renderlo numericamente stabile, razionalizziamo ottenendo sqrt(s^2 / (2 + 2*sqrt(1 - s^2)))
b = np.array([2])
s = np.array([1])
p = np.array([])
n = 50
for i in range(n):
    p = np.append(p, b[i] * s[i])
    b = np.append(b, 2 * b[i])
    s = np.append(s, np.sqrt((s[i]**2) / (2 * (1.0 + np.sqrt(1.0 - s[i]**2)))))
plt.title("Valori di p")
plt.plot(range(n), p, "r")
plt.hlines(math.pi, 0, n, colors="b", linestyles="--")
plt.show()

plt.title("Errore relativo")
plt.plot(range(n), e(p), "r")
plt.show()
```

## Esercizio 4   
Si consideri lo sviluppo in serie dell'esponenziale $e^x=1+x+\frac{1}{2!}x^2+\frac{1}{3!}x^3+\cdots =\sum_{k=0}^\infty \frac{x^k}{k!} $. Si calcoli un'approssimazione dell'esponenziale troncando ad N=100 termini la serie, per valori di x che variano da -10 a 10 incluso. Si calcoli  per ogni valore di x l'errore relativo del valore approssimato rispetto al valore esatto, ottenuto mediante la funzione exp della libreria numpy. Cosa si osserva? Pensare ad un algoritmo piÃ¹ stabile ricordando che $e^{-x}=\frac{1}{e^{|x|}}$.



```python
import numpy as np
import math
import matplotlib.pyplot as plt

def exp(x, N):
    e = np.zeros_like(x)
    for k in range(N):
        e += np.pow(x, k) / math.factorial(k)
    return e

def exp_optimized(x, N):
    return np.where(x < 0, 1.0 / exp(np.abs(x), N), exp(x, N))

def show(label, x, e, exact_e):
    # plt.plot(x, e, "r", x, exact_e, "b--")
    # plt.title("Valori")
    # plt.legend([label, "e^x (numpy)"])
    # plt.show()
    
    err = lambda x, y: np.abs(x - y) / np.abs(y)
    plt.plot(x, err(e, exact_e), "r")
    # plt.yscale("log")
    plt.title("Errore relativo")
    plt.show()

N = 100
x = np.arange(-10.0, 11.0)
e = exp(x, N)
exact_e = np.exp(x)
show("e^x", x, e, exact_e)
e = exp_optimized(x, N)
show("e^x (optimized)", x, e, exact_e)
```

## Esercizio 5
Calcolare l'approssimazione della derivata prima di f(x) = sin(x) in x = 1
mediante l'algoritmo del rapporto incrementale (f(x + h) - f(x))/h per valori decrescenti
di h, confrontandolo con il valore fornito dalla funzione di libreria per f'(x)
mediante calcolo dell'errore relativo.

```python
import numpy as np
import matplotlib.pyplot as plt

f = lambda x: np.sin(x)
f1 = lambda x: np.cos(x)
def rapporto_incrementale(f, x, h):
    return (f(x + h) - f(x)) / h

x = 1.0
h = 10.0 ** -np.arange(0, 21.0)
der_esatta = f1(x)
der_approssimata = rapporto_incrementale(f, x, h)

plt.semilogx(h, der_approssimata, "r", h, np.full_like(der_approssimata, der_esatta), "b--")
plt.title("Derivata")
plt.legend(["rapporto incrementale", "derivata esatta"])
plt.xlabel("h")
plt.show()

err: lambda x, y: np.abs(x - y) / np.abs(y)

plt.loglog(h, err(der_approssimata, der_esatta), "r")
plt.title("Errore relativo")
plt.xlabel("h")
plt.show()
```

## Esercizio 6
Valutazione di un polinomio in un vettore di punti
-   Realizzare una function  che mediante lo schema di Horner  valuti un polinomio in un vettore di punti.
-   Realizzare una function  che utilizzi lâalgoritmo classico per valutare un polinomio in un vettore di punti.
-   Valutare   i seguenti polinomi, con i due metodi implementati nei punti precedenti,  sul vettore di punti specificato  e  rappresentarli graficamente. Confrontarne i tempi 

         p(x)=x^8-16x^7+112x^6-448x^5+1120x^4-1792x^3+1792x^2-1024x+256  da valutare su vettore di  1000000 punti equidistanti in [1.8,2.2]
         p(x)= x^11 â8 x^9 +28 x^6 â7 x^4 +12 x^3 â4 x +1   da valutare su vettore di 1000000  punti in [-5:5]
         
Cosa conferma questo esperimento?


```python
import numpy as np
import matplotlib.pyplot as plt

def valuta_polinomio(p, x):
    y = np.zeros_like(x)
    for i in range(p.size):
        y += p[i] * (x**i)
    return y

def valuta_horner(p, x):
    y = np.full_like(x, p[p.size - 1])
    for i in reversed(range(p.size - 1)):
        y = (y * x) + p[i]
    return y

p1 = np.array([256, -1024, 1792, -1792, 1120, -448, 112, -16, 1])
x1 = np.linspace(1.8, 2.2, 1000000)
p2 = np.array([1, -4, 0, 12, -7, 0, 28, 0, 0, -8, 0])
x2 = np.linspace(-5, 5, 1000000)

def mostra_grafico(label, p, x):
    plt.plot(x, valuta_polinomio(p, x), "r", x, valuta_horner(p, x), "b--")
    plt.title(label)
    plt.xlabel("x")
    plt.ylabel("p(x)")
    plt.legend(["normale", "horner"])
    plt.show()

mostra_grafico("p(x)=x^8-16x^7+112x^6-448x^5+1120x^4-1792x^3+1792x^2-1024x+256", p1, x1)
mostra_grafico("p(x)= x^11 â8 x^9 +28 x^6 â7 x^4 +12 x^3 â4 x +1", p2, x2)
```
